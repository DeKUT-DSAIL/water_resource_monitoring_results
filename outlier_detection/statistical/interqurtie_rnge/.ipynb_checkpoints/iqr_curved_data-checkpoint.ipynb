### Import the dataset (from 24th April 2022 to 17 May 2022) 

```python
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
plt.rcParams['figure.figsize'] = (13, 7)
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('curved_labelled_data.csv')
df[['time']] = df[['time']].apply(pd.to_datetime)
df.rename(columns = {'Unnamed: 0':'x_column' }, inplace = True)
```

#### inter quantile range

```python
import datetime as datetime
water_level = np.array(df['height']).reshape(-1, 1)
```

#### iqr

```python
length = np.array(df.height)
water_arrays = np.array_split(length, 54)
len(length)
```

```python
Q1 = df.height.quantile(0.25)
Q3 = df.height.quantile(0.75)
IQR = Q3 - Q1
```

```python
lower_limit = Q1 - IQR
upper_limit = Q3 + IQR
upper_limit
```

### iqr.

```python
sup_list = []
for array1 in water_arrays:
    height_list = array1
    sup_list.append(height_list)
#print(sup_list)
```

```python
clas = []
for array2 in sup_list:
    clusters = []
    for i in array2:
        if i <  np.quantile(array2, 0.25) - ((np.quantile(array2,0.75) - np.quantile(array2, 0.25))) :
            clusters.append(1)
        elif i > np.quantile(array2,0.75) + ((np.quantile(array2,0.75) - np.quantile(array2, 0.25))):
            clusters.append(1)
        else: 
            clusters.append(0)
        #=clusters.append(clusters)
    clas += clusters 
    #print(len(clusters)) 
print(len(clas))
```

```python
df['iqr_labels'] = clas
```

```python
df.to_csv ('iqr_metric_curved.csv', header=True,index = False)
```

### Model metrics

```python
from sklearn.metrics import accuracy_score, f1_score,precision_score, recall_score, confusion_matrix, roc_curve, auc, ConfusionMatrixDisplay
data_labels = list(df['data_label'])
iqr_labels = list(df['iqr_labels'])
```
