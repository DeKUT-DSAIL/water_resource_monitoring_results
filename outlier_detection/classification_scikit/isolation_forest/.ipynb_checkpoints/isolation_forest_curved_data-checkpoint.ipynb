### Import the dataset (from 24th April 2022 to 17 May 2022) 

```python
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
plt.rcParams['figure.figsize'] = (15, 9)
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('curved_labelled_data.csv')
df[['time']] = df[['time']].apply(pd.to_datetime)
df.rename(columns = {'Unnamed: 0':'x_column' }, inplace = True)
```

### Plotting the dataset 

```python
plt.rcParams['figure.figsize'] = (15, 9)
fig, ax = plt.subplots(1,1)
fig.patch.set_facecolor('white')
ax.scatter(df['time'],df['height'], c= df['data_label'], cmap='rainbow')
ax.grid(color = 'gray', linestyle = '--', linewidth = 0.8)
ax.set_title('Outliers and the main trend from 24th April to 17th May 2022', fontsize=18, weight='bold')
ax.set_xlabel('Time', fontsize=18, weight = 'bold')
ax.set_ylabel('Water level (m)', fontsize=18, weight = 'bold')
ax.set_xlim(['2022-04-24 00:00:00'],['2022-05-17 23:59:59'])
ax.set_ylim(0, 1.9)
ax.set_xticklabels(df['time'], fontsize=16, weight='bold')
ax.tick_params(axis='y', labelsize=16)
ax.tick_params(axis='x',which='major', labelsize=16)

## Define the date format
date_form = DateFormatter("%d-%m")
ax.xaxis.set_major_formatter(date_form)
ax.set(facecolor = "white")

plt.show()
```

### Isolation Forest Algorithm.
- Return the anomaly score of each sample using the IsolationForest algorithm
- The IsolationForest isolates observations by randomly selecting a feature and then randomly selecting a split value between the maximum and minimum values of the selected feature.

```python
from sklearn.ensemble import IsolationForest
import datetime as datetime
import numpy as np
water_level = np.array(df['height'])
```

### Anomaly detection 

```python
length = np.array(df.height)
water_arrays = np.array_split(length, 1)
```

### Isolation Forest Algorithm.

```python
sup_list = []
for array in water_arrays:
    water_model = IsolationForest(n_estimators=95, random_state=0).fit(array.reshape(-1, 1))
    iso_labels = water_model.predict(array.reshape(-1, 1))
    sup_list.append(iso_labels)
#print(sup_list)
```

```python
l = []
for i in sup_list:
    for ii in i.tolist():
        l.append(ii)        
len(l)
```

```python
iso_labels = l
df['iso_labels'] = iso_labels
df.to_csv ('iso_output_curved.csv', header=True,index = False)
```

```python
iso_list = list(df['iso_labels'])
clusters =[]
for i in iso_list:
    if i == 1:
        cluster = 0
    else:
        cluster = 1
    clusters.append(cluster) 
    p = clusters
```

```python
df['iso_labels'] = p
```

```python
df.to_csv ('iso_metric_curved.csv', header=True,index = False)
```
