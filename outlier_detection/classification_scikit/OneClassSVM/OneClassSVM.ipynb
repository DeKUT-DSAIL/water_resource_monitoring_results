### Import modules

```python
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
plt.rcParams['figure.figsize'] = (12, 7)
import pandas as pd
import numpy as np
```

### Import the labelled dataset and plot it 

```python
df = pd.read_csv('labelled_data_sample.csv')
df[['time']] = df[['time']].apply(pd.to_datetime)
df.rename(columns = {'Unnamed: 0':'x_column' }, inplace = True)
```

```python
plt.rcParams['figure.figsize'] = (12, 7)
fig, ax = plt.subplots(1,1)
fig.patch.set_facecolor('white')
ax.scatter(df['time'],df['height'], c= df['datapoint_label'], cmap='rainbow')
ax.grid(color = 'gray', linestyle = '--', linewidth = 0.8)
ax.set_title('Outliers and the main trend from 6th March to 26th March 2022', fontsize=18)
ax.set_xlabel('Time', fontsize=18, weight = 'bold')
ax.set_ylabel('Water level (m)', fontsize=18, weight = 'bold')
ax.set_xlim(['2022-03-06 00:00:00'],['2022-03-26 23:59:59'])
ax.set_ylim(0, 1.4)
ax.set_xticklabels(df['time'], fontsize=16, weight='bold')
ax.tick_params(axis='y', labelsize=16)
ax.tick_params(axis='x',which='major', labelsize=16)

## Define the date format
date_form = DateFormatter("%d-%m")
ax.xaxis.set_major_formatter(date_form)
ax.set(facecolor = "white")

plt.show()
```

### OneClassSVM
- Semi- Unsupervised Outlier Detection using the Local Outlier Factor (LOF).
- Estimate the support of a high-dimensional distribution.

```python
from sklearn.svm import OneClassSVM
import datetime as datetime
def date_iteration(start_date, end_date):
    df_time = df.set_index('time')
    df_loc = df_time[start_date : end_date]
    df_plota =df_loc.reset_index()
    df_plota[['time']] = df_plota[['time']].apply(pd.to_datetime)
    num_of_clusters = 2
    water_level = np.array(df_plota['height']).reshape(-1, 1)
    SVM_labels_cc = OneClassSVM(kernel= 'linear', gamma='scale', nu=0.981).fit_predict(water_level)
    SVM_labels = list(SVM_labels_cc)
    return SVM_labels 
```

```python
list_1  = date_iteration ('2022-03-06 00:00:00', '2022-03-26 23:59:59') 

final_list = list_1 
```

### SVM labels on the dataframe 

```python
df['SVM_labels'] = final_list
df.to_csv ('SVM_output.csv', header=True,index = False)
```

### SVM  Output plot 

```python
SVM_list = list(df['SVM_labels'])
clusters =[]
for i in SVM_list:
    if i == -1:
        cluster = 0
    else:
        cluster = 1
    clusters.append(cluster) 
    p = clusters
```

```python
df['SVM_index'] = p
df2 = df.drop( ['SVM_labels','x_column' ], axis=1)
```

```python
df2.to_csv ('SVM_output_metric.csv', header=True,index = False)
```

```python
temp_df = df2.loc[(df2['SVM_index'] == 0)]
df3 = temp_df.set_index('time')
df4 = df3.reset_index()
```

```python
temp_df1 = df2.loc[(df2['SVM_index'] == 1)]
df5 = temp_df1.set_index('time')
df6 = df5.reset_index()
```

```python
plt.rcParams['figure.figsize'] = (11, 7)
fig, ax = plt.subplots(1,1)
fig.patch.set_facecolor('white')
Normal = ax.scatter(df4['time'],df4['height'], c ="black", marker="o", s = 10**1.25)
Outlier = ax.scatter(df6['time'],df6['height'], c ="red",marker ="o", s = 10**1.25)
ax.grid(color = 'gray', linestyle = '--', linewidth = 0.8)
ax.set_title('OneClassSVM classification output for the uniform data', fontsize=18, weight='bold')
ax.set_xlabel('Time', fontsize=18, weight = 'bold')
ax.set_ylabel('Water level (m)', fontsize=18, weight = 'bold')
ax.set_xlim(['2022-03-06 00:00:00'],['2022-03-26 23:59:59'])
ax.set_ylim(0, 1.4)
ax.set_xticklabels(df['time'], fontsize=16, weight='bold')
ax.tick_params(axis='y', labelsize=16)
ax.tick_params(axis='x',which='major', labelsize=16)

## Define the date format
date_form = DateFormatter("%d-%m")
ax.xaxis.set_major_formatter(date_form)
ax.legend(["Normal", "Outlier"], loc ="upper left", fancybox=True,facecolor='#01FFFF',prop={'size': 15,  'style': 'normal'})
ax.set(facecolor = "white")
plt.savefig('oneclass_normal.png', dpi=450, orientation='portrait', bbox_inches='tight', facecolor='w',edgecolor='b',)
plt.show()
```

### Model metrics 

```python
from sklearn.metrics import accuracy_score,f1_score, precision_score, recall_score, confusion_matrix, roc_curve, auc, ConfusionMatrixDisplay
data_labels = list(df2['datapoint_label'])
SVM_labels = list(df2['SVM_index'])
```

```python
c_matrix = confusion_matrix (data_labels,SVM_labels)
disp = ConfusionMatrixDisplay(c_matrix )
disp.plot(cmap=plt.cm.Blues,values_format='g',)
plt.xlabel('Predicted label', weight = 'bold')
plt.ylabel('True Label', weight = 'bold')


plt.savefig('con1.png', dpi=450, orientation='portrait', bbox_inches='tight', facecolor='w',edgecolor='b',)
plt.show()
```

```python
data_accuracy = accuracy_score (data_labels,SVM_labels)
data_accuracy
```

```python
data_precision = precision_score (data_labels,SVM_labels)
data_precision
```

```python
data_recall_score = recall_score (data_labels,SVM_labels)
data_recall_score
```

```python
data_f1_score = f1_score(data_labels,SVM_labels)
data_f1_score
```

```python
x = ['Accuracy', 'Precision', 'Recall', 'F1_score']
y = [data_accuracy, data_precision, data_recall_score,data_f1_score ]
```

```python
plt.rcParams['figure.figsize'] = (6, 4)
fig, ax = plt.subplots(1,1)
fig.patch.set_facecolor('white')
ax.bar(x, y, width=0.5)
ax.grid(color = 'gray', linestyle = '--', linewidth = 0.8)
ax.set_title('Performance metric score (LOF)', fontsize=14, weight='bold')
ax.set_xlabel('Metric', fontsize=14, weight = 'bold')
ax.set_ylabel('Metric value', fontsize=15, weight = 'bold')
ax.set_ylim(0, 1.1)
ax.set_xticklabels(x, fontsize=12, weight='bold')
ax.tick_params(axis='y', labelsize=14)
ax.tick_params(axis='x',which='major', labelsize=14)

## Define the date format
ax.set(facecolor = "white")
plt.savefig('bar1.png', dpi=450, orientation='portrait', bbox_inches='tight', facecolor='w',edgecolor='b',)
plt.show()
```
