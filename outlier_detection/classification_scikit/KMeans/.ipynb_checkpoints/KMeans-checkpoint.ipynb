### Import modules

```python
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter

plt.rcParams['figure.figsize'] = (12, 8)
import pandas as pd
import numpy as np
```

### Import the labelled dataset and plot it 

```python
df = pd.read_csv('labelled_data_sample.csv')
df[['time']] = df[['time']].apply(pd.to_datetime)
df.rename(columns = {'Unnamed: 0':'x_column' }, inplace = True)
```

```python
plt.rcParams['figure.figsize'] = (12, 8)
fig, ax = plt.subplots(1,1)
fig.patch.set_facecolor('white')
ax.scatter(df['time'],df['height'], c= df['datapoint_label'], cmap='rainbow')
ax.grid(color = 'gray', linestyle = '--', linewidth = 0.8)
ax.set_title('Outliers and the main trend from 6th March to 26th March 2022', fontsize=18)
ax.set_xlabel('Time', fontsize=18, weight = 'bold')
ax.set_ylabel('Water level (m)', fontsize=18, weight = 'bold')
ax.set_xlim(['2022-03-06 00:00:00'],['2022-03-26 23:59:59'])
ax.set_ylim(0, 1.4)
ax.set_xticklabels(df['time'], fontsize=16, weight='bold')
ax.tick_params(axis='y', labelsize=16)
ax.tick_params(axis='x',which='major', labelsize=16)

## Define the date format
date_form = DateFormatter("%d-%m")
ax.xaxis.set_major_formatter(date_form)
ax.set(facecolor = "white")

plt.show()
```

### KMeans clustering
- Exploratory Data Analysis (Get to know the distribution of the datapoints)
- To establish datapoint clusters
- Get rid of the unwanted clusters
- Number of clusters can be settled on by visualizing the data. (Alogrithms such as the silhouette_score and the elbow method can be used if the dataset is big or the the user is not familiar with data)

```python
from sklearn.cluster import KMeans

def date_iteration(start_date, end_date):
    df_time = df.set_index('time')
    df_loc = df_time[start_date : end_date]
    df_plota =df_loc.reset_index()
    df_plota[['time']] = df_plota[['time']].apply(pd.to_datetime)
    num_of_clusters = 2
    water_level = np.array(df_plota['height']).reshape(-1, 1)
    water_level_pred = KMeans(n_clusters = num_of_clusters, random_state = 123).fit_predict(water_level)
    water_level_pred2 = list(water_level_pred)
    return water_level_pred2
```

```python
list_1  = date_iteration ('2022-03-06 00:00:00', '2022-03-09 23:59:59') 
list_2  = date_iteration ('2022-03-10 00:00:00', '2022-03-14 23:59:59')
list_3  = date_iteration ('2022-03-15 00:00:00', '2022-03-17 23:59:59')
list_4  = date_iteration ('2022-03-18 00:00:00', '2022-03-20 23:59:59')
list_5  = date_iteration ('2022-03-21 00:00:00', '2022-03-23 23:59:59')
list_6  = date_iteration ('2022-03-24 00:00:00', '2022-03-26 23:59:59')
```

```python
final_list = list_1+list_2 +list_3+list_4+list_5+list_6    
```

```python
surprise_list = final_list
clusters =[]
for i in surprise_list:
    if i == 0:
        cluster = 0
    else:
        cluster = 1
    clusters.append(cluster) 
    p = clusters
```

```python
df['model_label'] = p
```

```python
temp_df = df.loc[(df['model_label'] == 0)]
df3 = temp_df.set_index('time')
df4 = df3.reset_index()
```

```python
temp_df1 = df.loc[(df['model_label'] == 1)]
df5 = temp_df1.set_index('time')
df6 = df5.reset_index()
```

### Plotting the clusters 

```python
plt.rcParams['figure.figsize'] = (12, 8)
fig, ax = plt.subplots(1,1)
fig.patch.set_facecolor('white')
Normal = ax.scatter(df4['time'],df4['height'], c ="black", marker="o", s = 10**1.35)
Outlier = ax.scatter(df6['time'],df6['height'], c ="red",marker ="o", s = 10**1.35)
ax.grid(color = 'gray', linestyle = '--', linewidth = 1.5)
ax.set_title('KMeans classification output for the uniform data', fontsize=18, weight='bold')
ax.set_xlabel('Time', fontsize=18, weight = 'bold')
ax.set_ylabel('Water level (m)', fontsize=18, weight = 'bold')
ax.set_xlim(['2022-03-06 00:00:00'],['2022-03-26 23:59:59'])
ax.set_ylim(0, 1.4)
ax.set_xticklabels(df['time'], fontsize=16, weight='bold')
ax.tick_params(axis='y', labelsize=16)
ax.tick_params(axis='x',which='major', labelsize=16)

## Define the date format
date_form = DateFormatter("%d-%m")
ax.xaxis.set_major_formatter(date_form)
ax.legend(["Normal", "Outlier"], loc ="upper left", fancybox=True,facecolor='#01FFFF',prop={'size': 15,  'style': 'normal'})
ax.set(facecolor = "white")

plt.savefig('kmeans_normal.png', dpi=450, orientation='portrait', bbox_inches='tight', facecolor='w',edgecolor='b',)
plt.show()
```

### Model metrics 

```python
from sklearn.metrics import accuracy_score, f1_score,precision_score, recall_score, confusion_matrix, roc_curve, auc, ConfusionMatrixDisplay
data_labels = list(df['datapoint_label'])
kmeans_labels = list(df['kmeans_labels'])
```

```python
c_matrix = confusion_matrix (data_labels,kmeans_labels)
disp = ConfusionMatrixDisplay(c_matrix )
disp.plot(cmap=plt.cm.Blues,values_format='g',)
plt.savefig('con1.png', dpi=450, orientation='portrait', bbox_inches='tight', facecolor='w',edgecolor='b',)
plt.show()
```

```python
data_accuracy = accuracy_score (data_labels,kmeans_labels)
data_accuracy
```

```python
data_precision = precision_score (data_labels,kmeans_labels)
data_precision
```

```python
data_recall_score = recall_score (data_labels,kmeans_labels)
data_recall_score
```

```python
data_f1_score = f1_score (data_labels,kmeans_labels)
data_f1_score
```

```python
x = ['Accuracy', 'Precision', 'Recall', 'F1_score']
y = [data_accuracy, data_precision, data_recall_score,data_f1_score ]
```

```python
plt.rcParams['figure.figsize'] = (6, 4)
fig, ax = plt.subplots(1,1)
fig.patch.set_facecolor('white')
ax.bar(x, y, width=0.5)
ax.grid(color = 'gray', linestyle = '--', linewidth = 0.8)
ax.set_title('Performance metric score', fontsize=14, weight='bold')
ax.set_xlabel('Metric', fontsize=14, weight = 'bold')
ax.set_ylabel('Metric value', fontsize=15, weight = 'bold')
ax.set_ylim(0, 1.1)
ax.set_xticklabels(x, fontsize=12, weight='bold')
ax.tick_params(axis='y', labelsize=14)
ax.tick_params(axis='x',which='major', labelsize=14)

## Define the date format
ax.set(facecolor = "white")
plt.savefig('bar1.png', dpi=450, orientation='portrait', bbox_inches='tight', facecolor='w',edgecolor='b',)
plt.show()
```
