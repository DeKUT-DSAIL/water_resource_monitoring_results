### Import the dataset (from 24th April 2022 to 17 May 2022) 

```python
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
plt.rcParams['figure.figsize'] = (15, 9)
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('curved_labelled_data.csv')
df[['time']] = df[['time']].apply(pd.to_datetime)
df.rename(columns = {'Unnamed: 0':'x_column' }, inplace = True)
```

### Plotting the dataset 

```python
plt.rcParams['figure.figsize'] = (12, 9)
fig, ax = plt.subplots(1,1)
fig.patch.set_facecolor('white')
ax.scatter(df['time'],df['height'], c= df['data_label'], cmap='rainbow')
ax.grid(color = 'gray', linestyle = '--', linewidth = 0.8)
ax.set_title('Outliers and the main trend from 24th April to 17th May 2022', fontsize=18, weight='bold')
ax.set_xlabel('Time', fontsize=18, weight = 'bold')
ax.set_ylabel('Water level (m)', fontsize=18, weight = 'bold')
ax.set_xlim(['2022-04-24 00:00:00'],['2022-05-17 23:59:59'])
ax.set_ylim(0, 1.9)
ax.set_xticklabels(df['time'], fontsize=16, weight='bold')
ax.tick_params(axis='y', labelsize=16)
ax.tick_params(axis='x',which='major', labelsize=16)

## Define the date format
date_form = DateFormatter("%d-%m")
ax.xaxis.set_major_formatter(date_form)
ax.set(facecolor = "white")

plt.savefig('sample_photo_curved.png', dpi=450, orientation='portrait', bbox_inches='tight', facecolor='w',edgecolor='b',)
plt.show()
```

### KMeans clustering
- Exploratory Data Analysis (Get to know the distribution of the datapoints)
- To establish datapoint clusters
- Get rid of the unwanted clusters
- Number of clusters can be settled on by visualizing the data. (Alogrithms such as the silhouette_score and the elbow method can be used if the dataset is big or the the user is not familiar with data)

```python
from sklearn.cluster import KMeans
import datetime as datetime
import numpy as np
water_level = np.array(df['height'])
```

### Anomaly detection 

```python
length = np.array(df.height)
length2 = np.array(df.height)
water_arrays = np.array_split(length, 25)
```

### KMeans water level data fitting 

```python
sup_list = []
for array in water_arrays:
    water_level_pred = KMeans(n_clusters = 2, random_state = 123 ).fit_predict(array.reshape(-1, 1))
    sup_list.append(water_level_pred)
#print(sup_list)
```

```python
l = []
for i in sup_list:
    for ii in i.tolist():
        l.append(ii)        
len(l)
```

```python
kmeans_labels = l
df['kmeans_labels'] = kmeans_labels
df.to_csv ('kmeans_output_curved.csv', header=True,index = False)
```

```python
surprise_list = kmeans_labels
clusters =[]
for i in surprise_list:
    if i == 0:
        cluster = 0
    else:
        cluster = 1
    clusters.append(cluster) 
    p = clusters
```

```python
df['kmeans_labels'] = p
```

```python
df.to_csv ('kmeans_output_metric_curved.csv', header=True,index = False)
```

```python
temp_df = df.loc[(df['kmeans_labels'] == 0)]
df3 = temp_df.set_index('time')
df4 = df3.reset_index()
```

```python
temp_df1 = df.loc[(df['kmeans_labels'] == 1)]
df5 = temp_df1.set_index('time')
df6 = df5.reset_index()
```

```python
plt.rcParams['figure.figsize'] = (12, 8)
fig, ax = plt.subplots(1,1)
fig.patch.set_facecolor('white')
Normal = ax.scatter(df4['time'],df4['height'], c ="black", marker="o", s = 10**1.35)
Outlier = ax.scatter(df6['time'],df6['height'], c ="red",marker ="o", s = 10**1.35)
ax.grid(color = 'gray', linestyle = '--', linewidth = 1.3)
ax.set_title('KMeans classification output for the non-uniform data', fontsize=18, weight='bold')
ax.set_xlabel('Time', fontsize=18, weight = 'bold')
ax.set_ylabel('Water level (m)', fontsize=18, weight = 'bold')
ax.set_xlim(['2022-04-26 00:00:01'],['2022-05-17 23:59:59'])
ax.set_ylim(0,1.9)
ax.set_xticklabels(df['time'], fontsize=16, weight='bold')
ax.tick_params(axis='y', labelsize=16)
ax.tick_params(axis='x',which='major', labelsize=16)

## Define the date format
date_form = DateFormatter("%d-%m")
ax.xaxis.set_major_formatter(date_form)
ax.legend(["Normal", "Outlier"], loc ="upper left", fancybox=True,facecolor='#01FFFF',prop={'size': 15,  'style': 'normal'})
ax.set(facecolor = "white")
plt.savefig('kmeans_curved.png', dpi=450, orientation='portrait', bbox_inches='tight', facecolor='w',edgecolor='b',)
plt.show()
```
