### Local Outlier Factor
- Unsupervised Outlier Detection using the Local Outlier Factor (LOF).
- It measures the local deviation of the density of a given sample with respect to its neighbors.

```python
from sklearn.neighbors import LocalOutlierFactor
import datetime as datetime
def date_iteration(start_date, end_date):
    df_time = df.set_index('time')
    df_loc = df_time[start_date : end_date]
    df_plota =df_loc.reset_index()
    df_plota[['time']] = df_plota[['time']].apply(pd.to_datetime)
    num_of_clusters = 2
    water_level = np.array(df_plota['height']).reshape(-1, 1)
    lof_labels_cc = LocalOutlierFactor(n_neighbors = 14, metric='euclidean').fit_predict(water_level)
    lof_labels = list(lof_labels_cc)
    return lof_labels 
```

### LOF water level data fitting 

```python
list_1  = date_iteration ('2022-03-06 00:00:00', '2022-03-07 23:59:59') 
list_2  = date_iteration ('2022-03-08 00:00:00', '2022-03-09 23:59:59')
list_3  = date_iteration ('2022-03-10 00:00:00', '2022-03-11 23:59:59')
list_4  = date_iteration ('2022-03-12 00:00:00', '2022-03-13 23:59:59')
list_5  = date_iteration ('2022-03-14 00:00:00', '2022-03-15 23:59:59')
list_6  = date_iteration ('2022-03-16 00:00:00', '2022-03-17 23:59:59')
list_7  = date_iteration ('2022-03-18 00:00:00', '2022-03-19 23:59:59')
list_8  = date_iteration ('2022-03-20 00:00:00', '2022-03-21 23:59:59')
list_9  = date_iteration ('2022-03-22 00:00:00', '2022-03-23 23:59:59')
list_10  = date_iteration ('2022-03-24 00:00:00', '2022-03-26 23:59:59')
final_list = list_1 + list_2 + list_3 + list_4 + list_5 + list_6 + list_7 + list_8 + list_9 + list_10
```

### LOF labels on the dataframe 

```python
lof_labels = final_list
df['lof_labels'] = lof_labels
df.to_csv ('lof_output.csv', header=True,index = False)
```

### Model metrics 

```python
from sklearn.metrics import accuracy_score, f1_score,precision_score, recall_score, confusion_matrix, roc_curve, auc, ConfusionMatrixDisplay
data_labels = list(df2['datapoint_label'])
lof_labels = list(df2['lof_index'])
```

```python
c_matrix = confusion_matrix (data_labels,lof_labels)
disp = ConfusionMatrixDisplay(c_matrix )
disp.plot(cmap=plt.cm.Blues,values_format='g',)
plt.xlabel('Predicted label', weight = 'bold')
plt.ylabel('True Label', weight = 'bold')


plt.savefig('con1.png', dpi=450, orientation='portrait', bbox_inches='tight', facecolor='w',edgecolor='b',)
plt.show()
```

```python
data_accuracy = accuracy_score (data_labels,lof_labels)
data_accuracy
```

```python
data_recall_score = recall_score (data_labels,lof_labels)
data_recall_score
```

```python
data_f1_score = f1_score (data_labels,lof_labels)
data_f1_score
```
